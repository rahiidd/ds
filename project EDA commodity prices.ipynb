```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
commodity_df = pd.read_csv('commodity_prices.csv')
```

```python
commodity_df
```

```python
commodity_df.isna().sum()
```

```python
commodity_df = commodity_df[['Year', 'Maize', 'Rice', 'Wheat']]
```

```python
commodity_df
```

```python
# SUMMARIZING
mean = commodity_df[['Maize', 'Rice', 'Wheat']].mean()
median = commodity_df[['Maize', 'Rice', 'Wheat']].median()
mode = commodity_df[['Maize','Rice','Wheat']].mode().values[0]
variance = np.var(commodity_df[['Maize','Rice','Wheat']])
std_deviation = np.std(commodity_df[['Maize','Rice','Wheat']])
```

```python
print('MEAN\n',mean)
print('MEDIAN\n',median)
print('MODE\n',mode)
print('VARIANCE\n',variance)
print('STD DEVIATION\n',std_deviation)
```
